```python
import pandas as pd
import sqlite3
```

```python
df=pd.read_csv('/content/Car Sales.xlsx - car_data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
rename_map = {
    "Car_id": "car_id",
    "Date": "date",
    "Customer Name": "customer_name",
    "Gender": "gender",
    "Annual Income": "annual_income",
    "Dealer_Name": "dealer_name",
    "Company": "company",
    "Model": "model",            # (you had 'mmodel'; assuming 'model' is intended)
    "Engine": "engine",
    "Transmission": "transmission",
    "Color": "color",
    "Price ($)": "price",
    "Dealer_No": "dealer_no",
    "Body Style": "body_style",
    "Phone": "phone",
    "Dealer_Region": "dealer_region",
}

df = df.rename(columns=rename_map)
```

```python
import prettytable
prettytable.DEFAULT='DEFAULT'
%load_ext sql
conn=sqlite3.connect('carsales.db')
cor=conn.cursor()
%sql sqlite:///carsales.db
```

```python
df.to_sql('car_sales',conn,if_exists='replace',index=False)
```

```python
%sql SELECT * FROM car_sales LIMIT 5
```

```python
%sql SELECT SUM(price) FROM car_sales;
```

```python
%sql SELECT COUNT(*) FROM car_sales;
```

```python
%%sql SELECT strftime('%Y-%m', "date") AS month,
COUNT(*) AS units_sold,
SUM("price") AS revenue,
AVG("price") AS avg_selling_price
FROM car_sales
GROUP BY month
ORDER BY month;
```

```python
%%sql SELECT dealer_name, dealer_region,
COUNT(*) AS units,
SUM("price") AS revenue,
AVG("price") AS asp
FROM car_sales
GROUP BY dealer_name, dealer_region
ORDER BY revenue DESC, units DESC;
```

```python
%%sql SELECT body_style,
COUNT(*) AS units,
AVG("price") AS avg_price,
MIN("price") AS min_price,
MAX("price") AS max_price
FROM car_sales
GROUP BY body_style
ORDER BY units DESC;
```

```python
%%sql SELECT model,dealer_region,
COUNT(*) AS units,
SUM("price") AS revenue,
AVG("price") AS asp
FROM car_sales
GROUP BY model, dealer_region
ORDER BY units DESC, revenue DESC;
```

```python
%%sql SELECT gender,dealer_region,
SUM(CASE WHEN "Transmission"='Auto' THEN 1 ELSE 0 END) AS automatic_units,
SUM(CASE WHEN "Transmission"='Manual' THEN 1 ELSE 0 END) AS manual_units,
ROUND(100.0 * SUM(CASE WHEN "Transmission"='Auto' THEN 1 ELSE 0 END) / COUNT(*), 1) AS automatic_share_pct
FROM car_sales
GROUP BY gender, dealer_region
ORDER BY automatic_share_pct DESC;
```

```python
%%sql SELECT company,engine,
COUNT(*) AS units,
AVG("price") AS avg_price
FROM car_sales
GROUP BY company, engine
ORDER BY units DESC, avg_price DESC;
```
